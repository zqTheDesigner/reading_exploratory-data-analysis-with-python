# Generating pairplots on two variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

penguins_data = pd.read_csv("data/penguins_size.csv")
penguins_data = penguins_data[['species', 'culmen_length_mm', 'body_mass_g', 'sex']]

sns.pairplot(data=penguins_data)
```

```python
plt.figure(figsize = (12,6))

ax = sns.barplot(data = penguins_data, x = 'species', y='culmen_length_mm', estimator = np.median)

ax.set_xlabel('Species', fontsize=15)
ax.set_ylabel('Culmen_length_mm', fontsize=15)

ax.set_title('Bivariate analysis of Culment Length and Species', fontsize=20)
```

```python
ax = sns.boxplot(penguins_data, x='species', y='culmen_length_mm')
```

```python
penguins_data_male = penguins_data.loc[penguins_data['sex'] == 'MALE', :]
penguins_data_female = penguins_data.loc[penguins_data['sex'] == 'FEMALE', :]

ax = sns.histplot(data=penguins_data_male, x='culmen_length_mm', alpha=0.5, color='blue')
ax = sns.histplot(data=penguins_data_female, x='culmen_length_mm', alpha=0.5, color='red')

ax.set_title('Bivariate analysis of Culmen Length and Sex')

plt.legend(['Male Penguins', 'Female Penguind'], loc="upper right")
```
